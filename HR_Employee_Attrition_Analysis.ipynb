```python
import pandas as pd

df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")

df.head()
```

```python
attrition_rate = df['Attrition'].value_counts(normalize=True) * 100
attrition_rate
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
sns.countplot(data=df, x='Attrition')
plt.title("Overall Attrition Count")
plt.show()
```

```python
plt.figure()
sns.countplot(data=df, x='Department', hue='Attrition')
plt.xticks(rotation=45)
plt.title("Attrition by Department")
plt.show()
```

```python
plt.figure()
sns.countplot(data=df, x='OverTime', hue='Attrition')
plt.title("Attrition vs Overtime")
plt.show()
```

```python
plt.figure()
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome')
plt.title("Attrition vs Monthly Income")
plt.show()
```

```python
plt.figure()
sns.histplot(data=df, x='Age', hue='Attrition', bins=30)
plt.title("Age Distribution by Attrition")
plt.show()
```

Key Findings:
1. Overall attrition rate is around 16%.
2. Sales department has highest attrition.
3. Overtime significantly increases attrition risk.
4. Lower income employees tend to leave more.
5. Younger employees are more likely to resign.

Business Recommendation:
- Reduce overtime burden
- Improve salary structure for junior employees
- Introduce retention programs in Sales department
